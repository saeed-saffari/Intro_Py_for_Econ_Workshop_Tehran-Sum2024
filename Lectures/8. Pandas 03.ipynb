## 3. Extra Work

```python
price_AAPL = data.DataReader('AAPL', start, end)
```

```python
price_AAPL.head()
```

```python
price_AAPL['daily_diffrence'] = price_AAPL['Adj Close'].diff()
price_AAPL.head()
```

```python
price_AAPL['signal'] = np.where(price_AAPL['daily_diffrence'] > 0, 1, 0)
price_AAPL.head(10)
```

```python
price_AAPL['position'] = price_AAPL['signal'].diff()
price_AAPL.head(10)
```

```python
price_AAPL[price_AAPL['position'] == 1.0].index
```

```python
price_AAPL[price_AAPL['position'] == 1.0]['Adj Close']
```

```python
plt.figure(figsize=(12,7))

plt.plot(price_AAPL['Adj Close'], c = 'r', lw = 2)
plt.plot(price_AAPL[price_AAPL['position'] == 1.0].index,
        price_AAPL[price_AAPL['position'] == 1.0]['Adj Close'], 'o', c='g', markersize = 7
        )

plt.plot(price_AAPL[price_AAPL['position'] == -1.0].index,
        price_AAPL[price_AAPL['position'] == -1.0]['Adj Close'], 'x', c='k', markersize = 7
        )

plt.grid()
plt.show()
```
