## Task 1

### Task 1.1: importing covid19 dataset 


```python
corona_dataset = pd.read_csv('time_series_covid19_confirmed_global.csv')
corona_dataset.head()
```

#### Let's check the shape of the dataframe

### Task 1.2: Delete the useless columns

```python
corona_dataset.drop(['Lat', "Long"], axis = 1, inplace=True)
```

### Task 1.3: Aggregating the rows by the country

```python
corona_dataset_aggregated = corona_dataset.groupby('Country/Region').sum(numeric_only=True)
corona_dataset_aggregated.head(15)
```

```python
corona_dataset_aggregated.shape
```

### Task 1.4: Visualizing data related to countries (for example China, Italy, Spain and Iran)

### Task 2.3: find maximum infection rate for all of the countries. 

```python
corona_dataset_aggregated.index
```

```python
country = list(corona_dataset_aggregated.index)
```

```python
#country
```

```python
max_infection = []

for c in country:
    max_infection.append(corona_dataset_aggregated.loc[c].diff().max())
```

```python
corona_dataset_aggregated['max_infection_rate'] = max_infection
```

```python
corona_dataset_aggregated.head()
```

### Task 2.4: create a new dataframe with only needed column 

```python
corona_2 = pd.DataFrame(corona_dataset_aggregated['max_infection_rate'])
corona_2.head()
```

### Task 5.1: Plotting GDP vs maximum Infection rate

```python
x = data['Logged GDP per capita']
y = data['max_infection_rate']

sns.scatterplot(x =x , y = np.log(y))
```

```python
plt.figure(figsize=(9,6))
sns.regplot(x =x, y = np.log(y))
plt.title('GDP per Capita VS Max infection rate', fontsize=15, fontname='Times New Roman')

plt.grid()
plt.show()
```

### Task 5.2: Plotting Social support vs maximum Infection rate

```python
x = data['Social support']
y = data['max_infection_rate']

sns.scatterplot(x =x , y = np.log(y))

plt.figure(figsize=(9,6))
sns.regplot(x =x, y = np.log(y))
plt.title('Social Support VS Max infection rate', fontsize=15, fontname='Times New Roman')

plt.grid()
plt.show()
```

### Task 5.3: Plotting Healthy life expectancy vs maximum Infection rate

```python
x = data['Healthy life expectancy']
y = data['max_infection_rate']

sns.scatterplot(x =x , y = np.log(y))

plt.figure(figsize=(9,6))
sns.regplot(x =x, y = np.log(y))
plt.title('Healthy life expectancy VS Max infection rate', fontsize=15, fontname='Times New Roman')

plt.grid()
plt.show()
```
